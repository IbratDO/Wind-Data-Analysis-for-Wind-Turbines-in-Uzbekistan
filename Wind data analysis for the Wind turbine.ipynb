Now we have got overview of the dataset. We need the average, min, max speed of wind. According to the information of the eia.gov, good places for wind turbines are where the annual average wind speed is at least 9 miles per hour(mph) or 14 kilometer per hour(km/h) for small wind turbines and 13 mph (20 km/h) for utility-scale turbines.

Now we have hourly, daily, monthly data about wind in Uzbekistan. It will be more easy and understandable if we change this info into daily data. It is easy to use underscores instead of brackets.

```python
df2 = df.set_axis(['time', 'windspeed_10m_kmh', 'windspeed_100m_kmh', 'windgusts_10m_kmh'], axis = 'columns')
df2
```

```python
daily_average = df2.groupby(['time'], sort = False).mean()
daily_min = df2.groupby(['time'], sort = False).min()
daily_max = df2.groupby(['time'], sort = False).max()
```

Now we have sorted the data into daily average, min and max.
Next it is more helpful how many days wind blows faster than 9 mph or 14 km/h because this is the minimum speed at which wind turbines can generate electricity.

```python
daily_average[(daily_average['windspeed_10m_kmh'] > 14) & (daily_average['windspeed_100m_kmh'] > 14)]
```

In the table above, wind's average speed is faster than 14 km/h in 697 days out of 1280. This means that if wind turbines are built, the effectiveness and efficiency of wind turbines are about 54% which is risky. 
But the table is based on historical data and it comprises wind speed occurring in whole Uzbekistan, not in the parts of it. Analysing the future trends of wind and its speed in the parts of Uzbekistan may be beneficial to build wind turbines, because different regions have different relief.
